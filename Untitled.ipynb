```python
def rank(df):
    return df.rank(axis=1, pct=True)


def scale(df):
    return df.div(df.abs().sum(axis=1), axis=0)


def log(df):
    return np.log1p(df)


def sign(df):
    return np.sign(df)


def power(df, exp):
    return df.pow(exp)


def ts_lag(df: pd.DataFrame, t: int = 1) -> pd.DataFrame:
    return df.shift(t)


def ts_delta(df, period=1):
    return df.diff(period)


def ts_sum(df: pd.DataFrame, window: int = 10) -> pd.DataFrame:
    return df.rolling(window).sum()


def ts_mean(df, window=10):
    return df.rolling(window).mean()


def ts_weighted_mean(df, period=10):
    return (df.apply(lambda x: WMA(x, timeperiod=period)))


def ts_std(df, window=10):
    return (df
            .rolling(window)
            .std())


def ts_rank(df, window=10):
    return (df
            .rolling(window)
            .apply(lambda x: x.rank().iloc[-1]))


def ts_product(df, window=10):
    return (df
            .rolling(window)
            .apply(np.prod))


def ts_min(df, window=10):
    return df.rolling(window).min()


def ts_max(df, window=10):
    return df.rolling(window).max()


def ts_argmax(df, window=10):
    return df.rolling(window).apply(np.argmax).add(1)


def ts_argmin(df, window=10):
    return (df.rolling(window)
            .apply(np.argmin)
            .add(1))


#def ts_corr(x, y, window):
#    res = Parallel(n_jobs=cpu_count())(delayed(rolling_apply)(dcor, window, x[col], y[col]) for col in x)
#    res = pd.DataFrame.from_dict(dict(zip(x.columns, res)))
#    res.index = x.index
#    res.columns.name = 'ticker'
#    return res

def ts_corr(x, y, window=10):
    res = x.rolling(window).corr(y)
    return res

def ts_cov(x, y, window=10):
    return x.rolling(window).cov(y)


```

```python
def get_mutual_info_score(returns, alpha, n=100000):
    df = pd.DataFrame({'y': returns, 'alpha': alpha}).dropna().sample(n=n)
    return mutual_info_regression(y=df.y, X=df[['alpha']])[0]
# Now, implement the alpha001 function
def alpha001(c, r):
    """
    Alpha001 from 101 Alphas Formula: (rank(ts_argmax(power(((returns < 0) ? ts_std(returns, 20) : close), 2.), 5)) -0.5)
    - Alpha001 is a ranking signal used to gauge relative performance.
    - Lower absolute Alpha001 values suggest better performance, while higher absolute values indicate worse performance.
    """
    # Replace closing price with 20-day standard deviation for negative returns
    c_copy = c.copy()  # Copy the close data
    c_copy[r < 0] = r.rolling(20).std()  # Apply 20-day rolling std for negative returns
    
    # Compute the squared values
    c_squared = np.power(c_copy, 2)
    
    # Find the index of the maximum value in a 5-day window
    ts_argmax_5 = c_squared.rolling(5).apply(np.argmax, raw=True)
    
    # Rank the results and multiply by -0.5 to generate the alpha signal
    alpha_signal = ts_argmax_5.rank(axis=1).mul(-0.5)
    
    return alpha_signal

def alpha002(o, c, v):
    """(-1 * ts_corr(rank(ts_delta(log(volume), 2)), rank(((close - open) / open)), 6))"""
    s1 = rank(ts_delta(log(v), 2))
    s2 = rank((c / o) - 1)
    alpha = -ts_corr(s1, s2, 6)
    return alpha.stack('ticker').swaplevel().replace([-np.inf, np.inf], np.nan)
```

```python
import yfinance as yf
import pandas as pd
import numpy as np

# Download historical data using Yahoo Finance
def download_data(ticker):
    stock_data = yf.download(ticker, period = '60d', interval= '15m')
    stock_data['Return'] = stock_data['Close'].pct_change()
    return stock_data

# Alpha 001: Momentum Indicator
def alpha_001(data):
    return data['Close'] - data['Close'].shift(1)

# Alpha 003: Reversal Signal
def alpha_003(data):
    return -(data['Close'] - data['Close'].shift(1))**2

# Alpha 008: Volatility Signal
def alpha_008(data):
    avg_price = (data['Close'] + data['High'] + data['Low']) / 3
    return data['Volume'].diff(2) * (1 - (avg_price.rank()))

# Alpha 018: RSI Based Alpha
def alpha_018(data):
    return data['Close'].rolling(window=5).std() + data['Close'].rolling(window=10).corr(data['Volume'])

# Alpha 036: Moving Average Convergence-Divergence (MACD) Signal
def alpha_036(data):
    short_sma = data['Close'].rolling(window=10).mean()
    long_sma = data['Close'].rolling(window=20).mean()
    return short_sma - long_sma

# Alpha 042: Trend Strength
def alpha_042(data):
    return (data['High'] * data['Low']).rank()

# Alpha 092: Relative Volume Indicator
def alpha_092(data):
    return (data['Close'].rank() - data['Volume'].rank()) / data['Close'].rank()

# Alpha 041: Moving Average Crossover Signal
def alpha_041(data):
    sma_20 = data['Close'].rolling(window=20).mean()
    return data['Close'] - sma_20

# Combine all Alphas into DataFrame
def calculate_alphas(data):
    data['Alpha_001'] = alpha_001(data)
    data['Alpha_003'] = alpha_003(data)
    data['Alpha_008'] = alpha_008(data)
    data['Alpha_018'] = alpha_018(data)
    data['Alpha_036'] = alpha_036(data)
    data['Alpha_042'] = alpha_042(data)
    data['Alpha_092'] = alpha_092(data)
    data['Alpha_041'] = alpha_041(data)
    return data

# Main function to get stock data and calculate alphas
ticker = "SPY"  # example: Apple Inc.
start = "2023-01-01"
end = "2024-10-08"
    
    # Download the stock data
stock_data = download_data(ticker)
    
    # Calculate alphas
alphas_data = calculate_alphas(stock_data).round(2)
    
    # Display the alphas
#alphas_data.tail(10)
alphas_data[['Close','Volume','Alpha_001', 'Alpha_003', 'Alpha_008', 'Alpha_018', 'Alpha_036', 'Alpha_042', 'Alpha_092', 'Alpha_041']].tail(50)
```
